```python
import pandas as pd
import matplotlib
import seaborn as sns
import scipy
from scipy import stats
```

```python
# Загрузить файл wines_data.csv
wine = pd.read_csv('wine_data.csv')
wine.head()
```

```python
cor = wine.corr()
cor
```

```python
matplotlib.pyplot.figure(figsize=(10, 10))
sns.heatmap(cor, annot=True, xticklabels=cor.columns.values, yticklabels=cor.columns.values)
```

```python
sns.pairplot(wine, hue='Class')
```

```python
# Визуализировать взаимосвязи между 3-4 переменными
```

```python
sns.jointplot(wine['Alcohol'], wine['Malic acid'])
```

```python
sns.jointplot(wine['Alcalinity of ash'], wine['Color intensity'])
```

```python
sns.jointplot(wine['Flavanoids'], wine['Nonflavanoid phenols'])
```

```python
sns.jointplot(wine['Color intensity'], wine['Flavanoids'])
```

```python
# Рассчитать t-test статистики для 3-4 переменных

stats.ttest_ind(wine['Total phenols'],wine['Flavanoids'])
```

```python
stats.ttest_ind(wine['Nonflavanoid phenols'],wine['Proanthocyanins'])
```

```python
stats.ttest_ind(wine['Malic acid'],wine['Ash'])
```

```python
# Для вин разных классов
sample_1 = wine[wine['Class'] == 1 ]
sample_2 = wine[wine['Class'] == 2 ]
sample_3 = wine[wine['Class'] == 3 ]

stats.ttest_ind(sample_1['Alcohol'],sample_2['Alcohol'])
```

```python
stats.ttest_ind(sample_3['Alcohol'],sample_2['Alcohol'])
```

```python
stats.ttest_ind(sample_1['Alcohol'],sample_3['Alcohol'])
```

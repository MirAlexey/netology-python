```python
import pandas as pd
import scipy
import seaborn as sns

from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew


```

```python
# 1.Загрузить файл auto.csv
data = pd.read_csv('auto.csv')
data.head()
```

```python
# BИнформация о датасете
data.info()
```

```python
data['engine-type'].value_counts()
```

```python
# Рассчитать основные статистики для переменных
# (среднее, медиана, мода, мин/макс, сред. отклонение)

#  С помощью встроенной фыункции describe
data.describe()
```

```python
# 3.Рассчитать и визуализировать корреляционную матрицу для переменных

# Рассчитать корреляционную матрицу
cor = data.corr()
cor
```

```python
# Визуализируем корреляционную матрицу с помощью метода heatmap библиотеки seaborn
sns.heatmap(cor, annot=True, xticklabels=cor.columns.values, yticklabels=cor.columns.values)
```

```python
# 4.Визуализировать распределение переменной ‘Price’, рассчитать показатели skewness и kurtosis

# Визуализация
sns.distplot(data['price'])
```

```python
# kurtosis (коэффициент эксцесса)
kurtosis(data['price'])
```

```python
# skewness (коэффициент ассиметрии)
skew(data['price'])
```

```python
# 5.Визуализировать несколько взаимосвязей между переменными

# Для отображения взаимосвязей добавим переменные отражающие не численные характеристики

# номер производителя (отсортирован по возрастанию средней цены)
make_num = {v:k for k,v in enumerate(data[['make','price']].groupby('make').mean().reset_index().sort_values('price')['make'])}
# тип кузова (отсортирован по возрастанию средней цены)
body_style_num = {v:k for k,v in enumerate(data[['body-style','price']].groupby('body-style').mean().reset_index().sort_values('price')['body-style'])}
# количество цилиндров 
num_of_cylinders = {'two': 2, 'three': 3,'four': 4, 'five': 5,'six': 6,'eight': 8, 'twelve': 12}

# добавляем в датафрейм численные данные
data['make_num'] = data.apply(lambda x: make_num[x['make']], axis=1)
data['body_style_num'] = data.apply(lambda x: body_style_num[x['body-style']], axis=1)
data['num_of_cylinders'] = data.apply(lambda x: num_of_cylinders[x['num-of-cylinders']], axis=1)
data.head()
```

```python
# для отображения зависимости используем метод pairplot библиотеки seaborn отображающий зависимости переменных между собой
# цветовая диференциация (куда же без неё) по типу производителя
sns.pairplot(data[['price','make_num', 'num_of_cylinders', 'horsepower', 'peak-rpm', 'body_style_num']], hue='make_num')
```

```python
# 6.С использованием t-test статистики проверить взаимосвязи между несколькими переменными
```

```python
# связь между ценами двух топовых производителей 
sample_1 = data[data['make'] == 'audi' ]
sample_2 = data[data['make'] == 'bmw' ]

stats.ttest_ind(sample_1['price'],sample_2['price'])
```

```python
# связь между ценами двух топовых производителей
sample_1 = data[data['make'] == 'mercedes-benz' ]
sample_2 = data[data['make'] == 'bmw' ]

stats.ttest_ind(sample_1['price'],sample_2['price'])
```

```python
# связь между мощностями двигателей с разными  максимальными оборотами
sample_1 = data[data['peak-rpm'] == 5000 ]
sample_2 = data[data['peak-rpm'] == 5500 ]

stats.ttest_ind(sample_1['horsepower'],sample_2['horsepower'])
```

```python
data['engine-type'].value_counts()
```

```python
# связь между мощностями двигателей разных типов
sample_1 = data[data['engine-type'] == 'ohcf' ]
sample_2 = data[data['engine-type'] == 'ohcv' ]

stats.ttest_ind(sample_1['horsepower'],sample_2['horsepower'])
```

```python
# связь между мощностями вигателей разных типов (самый распространенный и один из редких)
sample_1 = data[data['engine-type'] == 'ohc' ]
sample_2 = data[data['engine-type'] == 'rotor' ]

stats.ttest_ind(sample_1['horsepower'],sample_2['horsepower'])
```
